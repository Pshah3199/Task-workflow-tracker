```python
import pandas as pd

# Sample task workflow data
data = {
    "Task": [
        "Prepare Report", "Data Cleaning", "Client Follow-up",
        "Inventory Check", "Email Campaign", "System Update"
    ],
    "Assigned_To": [
        "Amit", "Neha", "Rahul", "Amit", "Neha", "Rahul"
    ],
    "Status": [
        "Completed", "In Progress", "Pending",
        "Completed", "In Progress", "Pending"
    ],
    "Days_Taken": [2, 4, 5, 1, 3, 6]
}

df = pd.DataFrame(data)
df
```

```python
status_count = df["Status"].value_counts()
status_count
```

```python
import matplotlib.pyplot as plt

status_count.plot(kind="bar")
plt.title("Task Status Overview")
plt.xlabel("Status")
plt.ylabel("Number of Tasks")
plt.show()
```

```python
workload = df["Assigned_To"].value_counts()
workload
```

```python
workload.plot(kind="bar")
plt.title("Tasks per Employee")
plt.xlabel("Employee")
plt.ylabel("Number of Tasks")
plt.show()
```

```python
delayed_tasks = df[df["Days_Taken"] > 4]
delayed_tasks
```
