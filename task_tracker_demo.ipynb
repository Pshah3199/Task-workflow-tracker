```python
# task_tracker_demo.ipynb

import csv
import os

FILENAME = "tasks.csv"

def add_task(task_name, priority, status="Pending"):
    file_exists = os.path.isfile(FILENAME)

    with open(FILENAME, "a", newline="") as file:
        writer = csv.writer(file)
        if not file_exists:
            writer.writerow(["Task Name", "Priority", "Status"])
        writer.writerow([task_name, priority, status])

def view_tasks():
    if not os.path.exists(FILENAME):
        print("No tasks found.")
        return

    with open(FILENAME, "r") as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)

# ---- DEMO RUN ----
add_task("Prepare report", "High")
add_task("Update dashboard", "Medium")
add_task("Send follow-up email", "Low", "In Progress")

print("Task Records:")
view_tasks()
```
